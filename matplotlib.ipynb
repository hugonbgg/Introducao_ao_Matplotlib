```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Criando os gráficos

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Plota o gráfico
#label é para definir a legenda
#Color é a cor, aceita HEX da cor tbm
#linewidth é a largura da linha
#marker são os marcadores dos intervalos/valores
#linestyle é o estilo da linha
plt.plot(x,y, label = '2x', color='red', linewidth = 2,
marker = '.', markersize = 20, markerfacecolor = 'green', markeredgecolor = 'black', linestyle = '--')

#Definição rápida de estilo (shorthand notation)
#'Marcador,linha,cor'
#plt.plot(x,y,'.--r' )

#Define o título, e os rótulos X e Y, bem como as fontes e tamanhos usados
plt.title('Meu primeiro gráfico', fontdict={'fontname': 'Comic Sans Ms', 'fontsize':20})
plt.xlabel('X Axis', fontdict={'fontname': 'Arial', 'fontsize':15})
plt.ylabel('Y Axis', fontdict={'fontname': 'Arial', 'fontsize':15})

#Define as escalas usadas no X e Y, as marcações
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#plota a legenda
plt.legend()
#Exibe o gráfico
plt.show()
```

### Criando um gráfico com duas linhas

```python
#criando uma array de 0 até 4 com intervalo de 0.5
x2 = np.arange(0,4.5,0.5)
x2
```

```python
#plotando o gráfico com duas linhas

#primeira linha:
plt.plot(x,y,'.--r', label = '2x' )
#segunda linha
plt.plot(x2, x2**2, label = 'X^2')

plt.legend()
plt.show()
```

```python
#plotando o gráfico com duas linhas e simulando uma projeção na plotagem

#alterando o tamanho do gráfico
plt.figure(figsize=(5,3), dpi = 300)

#primeira linha:
plt.plot(x,y,'.--r', label = '2x' )

#plotando os 6 primeiros valores da segunda linha
plt.plot(x2[:6], x2[:6]**2,'g', label = 'X^2')
#plotando os últimos valores da segunda linha
plt.plot(x2[5:], x2[5:]**2,'g--', label = 'X^2')

#plotando a legenda
plt.legend()

#Salvando o gráfico
plt.savefig('mygraph.png', dpi = 300)

#Exibindo o gráfico
plt.show()


```
